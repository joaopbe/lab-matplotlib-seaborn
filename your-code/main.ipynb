#### Plot both $y=x^2$ and $y=e^x$ in the same plot using normal and logarithmic scale.
**Hint**: Use `set_xscale` and `set_yscale`.

```python
fig, axs = plt.subplots(1, 2, figsize=(20, 10))

axs[0].plot(x, x**2, label='quadratic function', lw = 3, ls = '-.', color = 'blue')
axs[0].plot(x, np.exp(x) , label='exponential function', lw = 3, ls = '--', color = 'red')  

axs[1].plot(x, x**2, label='quadratic function', lw = 3, ls = '-.', color = 'blue')
axs[1].plot(x,np.exp(x) , label='exponential function', lw = 3, ls = '--', color = 'red')  

axs[1].set_xscale('log')
axs[1].set_yscale('log')

axs[0].set_xlabel('xlabel 1',fontsize=12, fontweight = 'roman')
axs[1].set_xlabel('xlabel 2',fontsize=12, fontweight = 'roman')

axs[0].set_ylabel('y label 1',fontsize=12, fontweight = 'roman')
axs[1].set_ylabel('y label 2',fontsize=12, fontweight = 'roman')

axs[0].set_title('Linear Scales',fontsize=14, fontweight = 'regular')
axs[1].set_title('Log Scales',fontsize=14, fontweight = 'regular')

plt.suptitle('Exponential versus Quadratic in linear and log scales', fontsize=16, fontweight = 'bold')
```

#### From the Fitbit data, we want to visually understand:

How the average number of steps change by month. Use the appropriate visualization to show the median steps by month. Is Fitbitter more active on weekend or workdays? All plots must be in the same jupyter notebook cell.

**Hints**:

* Use Months_encoded and Week_or Weekend columns.
* Use matplolib.pyplot object oriented API.
* Set your size figure to 12,4
* Explore plt.sca
* Explore plt.xticks
* Save your figures in a folder called `figures` in your repo. 

```python
# Calculating values
agg_months = fitbit.groupby('Months_encoded').median() [['Steps']]
agg_weeks = fitbit.groupby(['Work_or_Weekend']).median() [['Steps']]
agg_weeks

#Setting Figure
fig, axs = plt.subplots(1, 2, figsize=(12, 4))

# MEDIAN STEPS BY MONTH_ENCODED
plt.sca(axs[0])
plt.bar(agg_months.index, agg_months['Steps'], label='Per Month', alpha = 0.7)
plt.xticks(np.arange(1, 13, step=1))
plt.xlabel('Months',fontsize=10)
plt.ylabel('Median Steps',fontsize=10)


# MEDIAN STEPS BY WORK_OR_WEEKEND
plt.sca(axs[1])
plt.bar(agg_weeks.index, agg_weeks['Steps'], label='Per Month/Weekend', color = 'blue', alpha = 0.2)
plt.xlabel('Months',fontsize=10)
plt.ylabel('Median Steps',fontsize=10)
plt.ylim(9000,11000)
plt.xticks(agg_weeks.index, ('Weekends', 'Weekdays'))

#Saving figure
os.mkdir('../figures')
plt.savefig('../figures/Steps.png', dpi=200, format = 'png')
```

#### Use subplots and plot the distribution of the `Age`  with bins equal to 10, 20 and 50.

```python
fig, axs = plt.subplots(1, 3, figsize=(18, 5))
bins = [10, 20, 50]
ticks = [np.arange(0,90,10), np.arange(0,85,5),np.arange(0,85,5)]

axs[0].set_ylabel('# Samples',fontsize=10)

for ax, b, t in zip(axs,bins, ticks): 
    plt.sca(ax)
    titanic['Age'].hist(bins = b)
    ax.set_xlabel('Age',fontsize=10)
    ax.set_xticks(t)
    ax.set_title('Bins = %d' % b, fontsize = 10)

title = plt.suptitle('Age Distribution of Titanic Passengers', fontsize = 12)
```

#### How does the bin size affect your plot?

```python
"""
The bin size gives the number of equally spaced bins in the range os ages. As it increases, the ages are divided 
in progressively smaller bins. The consequence is we have more entries in the graph with each entry reporting 
lower values. In graphical terms we are trading granularity of the analysis for readability of the graphic
"""
```
